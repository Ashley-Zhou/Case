## 1. Ridge Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn import preprocessing

def Ridge_Regression_Test(equity_X, equity_y, rrr, cv, sheetname):
    
#     scaler = preprocessing.StandardScaler().fit(equity_X)
#     Equity_X = scaler.transform(equity_X)

    Equity_X_train, Equity_X_test, Equity_y_train, Equity_y_test = train_test_split(equity_X, equity_y, test_size=0.2, shuffle = True)
    print('Train and Test size:', len(Equity_X_train), len(Equity_X_test))

    # use Cross Validaiton to get the optimal penalization coefficien
    from sklearn.linear_model import RidgeCV
    ridgecv = RidgeCV(cv=cv, fit_intercept=False)
    ridgecv.fit(Equity_X_train, Equity_y_train)
    alpha_0 = ridgecv.alpha_

    #the ultimate alpha for Ridge Regression Model
    print('Penalizing Value: {}'.format(str(alpha_0)))

    #import machine learning model Ridge regression
    from sklearn.datasets import make_regression
    from sklearn.linear_model import Ridge

    #ridge regression
    ridge = Ridge(alpha=alpha_0)

    ridge.set_params(fit_intercept=False, normalize=True, random_state=rrr)

    ridge.fit(Equity_X_train, Equity_y_train)


    def pretty_print_linear(coefs, names = None, sort = False):
        if names == None:
            names = ['X%s' % x for x in range(len(coefs))]
        lst = zip(coefs, names)
        if sort:
            lst = sorted(lst, key = lambda x: -np.abs(x[0]))
        return '+'.join('%s * %s' % (round(coef, 3), name) for coef, name in lst)

    print ('Y = ' + str(ridge.intercept_) +' +  '+ pretty_print_linear(ridge.coef_, names = list(index.columns)))



    #get predicted y
    equity_y_pred = ridge.predict(equity_X)

    print("Training set score: {:.2f}".format(ridge.score(Equity_X_train, Equity_y_train))) #Training set score
    print("Test set score: {:.2f}".format(ridge.score(Equity_X_test, Equity_y_test))) #Test set score
    print("Total Coefficient Weight: {:.2f}".format(sum(ridge.coef_))) #total weight
    
    #plot weight of each coefficient
    plt.figure(figsize=(12,6))
    plt.plot(ridge.coef_ )
    plt.legend(['lasso regression coef with R_square: {%f}' %ridge.score(equity_X, equity_y)])
    plt.title('time series change comparison')
    plt.xticks(np.arange(len(index.columns)), list(index.columns), rotation=90)
    plt.show()
    
    #plot the actual&predicted performance
    plt.plot(equity_y, color = 'green', label = 'Actual Performance')
    plt.plot(equity_y_pred, color = 'black', label = 'Predicted Performance')
    plt.title('Performance')
    plt.xlabel('Time')
    plt.ylabel('Performance')
    plt.legend()
    plt.show()
    
    #calculate the cumulative price change start from 100
    equity_y_cum = 100*(equity_y +1).cumprod()
    equity_y_pred_cum = 100*(equity_y_pred +1).cumprod()
    
    #plot the cumulative actual&predicted performance
    plt.plot(equity_y_cum, color = 'green', label = 'Performance')
    plt.plot(equity_y_pred_cum, color = 'black', label = 'Predicted Performance')
    plt.title('VAMI Performance')
    plt.xlabel('Time')
    plt.ylabel('Performance')
    plt.legend()
    plt.show()
    
    p_equity = pd.read_excel('Case project.xlsx', sheet_name='Prediction', index_col=0)
    p_equity_imp = p_equity.loc[:, 'MSCI ACWI IMI Index (gd)':'BB US CMBS ex-AAA']
    p_equity_index = p_equity.iloc[:, 3:]
    p_equity_index = p_equity_index.drop(columns=['MSCI U.S. IMI Index (gross)','BB US CMBS ex-AAA TR Index Unhedged USD','Barclays US Aggregate Bond Index','MSCI World IMI ex-U.S. Index (net)'])
    p_equity_index

    p_equity_array = p_equity_index.values
    print(p_equity_array.shape)
    
    #get predicted value from the testing set
    ridge_pred_equity = ridge.predict(p_equity_array)
    
    ridge_pred_equity
    
    ridge_pred_equity_df = pd.DataFrame({sheetname: ridge_pred_equity})
    
    ridge_pred_equity_df.to_excel("output.xlsx",  sheet_name= sheetname)
```

## 2. Lasso Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn import preprocessing

def Lasso_Regression_Test(equity_X, equity_y, rrr, cv, sheetname):
    
#     scaler = preprocessing.StandardScaler().fit(equity_X)
#     Equity_X = scaler.transform(equity_X)

    Equity_X_train, Equity_X_test, Equity_y_train, Equity_y_test = train_test_split(equity_X, equity_y, test_size=0.2, shuffle=True)
    print('Train and Test size', len(Equity_X_train), len(Equity_X_test))

    # use Cross Validaiton to get the optimal penalization coefficient
    from sklearn.linear_model import LassoCV
    lassocv = LassoCV(cv=cv, max_iter = 1000000, fit_intercept=False)
    lassocv.fit(Equity_X_train, Equity_y_train)
    alpha_0 = lassocv.alpha_

    #the ultimate alpha for Lasso Regression Model
    print(str(alpha_0))

    #import machine learning model Lasso regression
    from sklearn.datasets import make_regression
    from sklearn.linear_model import Lasso

    #lasso regression
    lasso = Lasso(alpha=alpha_0)

    lasso.set_params(fit_intercept=False, normalize=True, random_state=rrr)

    lasso.fit(Equity_X_train, Equity_y_train)

    # lasso.predict()

    n = np.sum(lasso.coef_!= 0)
    print(n)

    def pretty_print_linear(coefs, names = None, sort = False):
        if names == None:
            names = ['X%s' % x for x in range(len(coefs))]
        lst = zip(coefs, names)
        if sort:
            lst = sorted(lst, key = lambda x: -np.abs(x[0]))
        return '+'.join('%s * %s' % (round(coef, 3), name) for coef, name in lst)

    print ('Y = ' + str(lasso.intercept_) +' +  '+ pretty_print_linear(lasso.coef_, names = list(index.columns)))

    print(lasso.sparse_coef_)

    print("Training set score: {:.2f}".format(lasso.score(Equity_X_train, Equity_y_train))) #Training set score
    print("Test set score: {:.2f}".format(lasso.score(Equity_X_test, Equity_y_test))) #Test set score
    print("Total Coefficient Weight: {:.2f}".format(sum(lasso.coef_))) #total weight

    #get predicted value from the testing set
    equity_y_pred_l = lasso.predict(equity_X)

    #plot weight of each coefficient
    plt.figure(figsize=(12,6))
    plt.plot(lasso.coef_ )
    plt.legend(['lasso regression coef with R_square: {%f}' %lasso.score(equity_X, equity_y)])
    plt.title('time series change comparison')
    plt.xticks(np.arange(len(index.columns)), list(index.columns), rotation=90)
    plt.show()

    #plot the actual&predicted performance
    plt.plot(equity_y, color = 'green', label = 'Actual Performance')
    plt.plot(equity_y_pred_l, color = 'black', label = 'Predicted Performance')
    plt.title('Performance')
    plt.xlabel('Time')
    plt.ylabel('Performance')
    plt.legend()
    plt.show()

    #calculate the cumulative performance of actual&predicted performance
    equity_y_cum = 100*(equity_y +1).cumprod()
    equity_y_pred_cum_l = 100*(equity_y_pred_l +1).cumprod()

    #plot the cumulative actual&predicted performance
    plt.plot(equity_y_cum, color = 'green', label = 'Performance')
    plt.plot(equity_y_pred_cum_l, color = 'black', label = 'Predicted Performance')
    plt.title('VAMI Performance')
    plt.xlabel('Time')
    plt.ylabel('Performance')
    plt.legend()
    plt.show()

    
    p_equity = pd.read_excel('Case project.xlsx', sheet_name='Prediction', index_col=0)
    p_equity_imp = p_equity.loc[:, 'MSCI ACWI IMI Index (gd)':'BB US CMBS ex-AAA']
    p_equity_index = p_equity.iloc[:, 3:]
    p_equity_index = p_equity_index.drop(columns=['MSCI U.S. IMI Index (gross)','BB US CMBS ex-AAA TR Index Unhedged USD','Barclays US Aggregate Bond Index','MSCI World IMI ex-U.S. Index (net)'])
    p_equity_index

    p_equity_array = p_equity_index.values
    print(p_equity_array.shape)
    
    #get predicted value from the testing set
    lasso_pred_equity = lasso.predict(p_equity_array)
    
    lasso_pred_equity
    
    lasso_pred_equity_df = pd.DataFrame({sheetname: lasso_pred_equity})

    lasso_pred_equity_df.to_excel("output.xlsx",  sheet_name= sheetname)
    
```
